3.3 Categorical Data Encoding

one hot encoding is a common way to encode categorical variables. <br />
number of columns is N-1, where N = number of categorical variables. <br />
categorical --> binary columns 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

plt.rcParams["figure.figsize"] = [8,6]
sns.set_style("darkgrid")

titanic_data = sns.load_dataset('titanic')

titanic_data.head()
```

selecting categorical variables sex, class and embark_town

```python
titanic_data = titanic_data[['sex','class','embark_town']]
titanic_data.head()
```

```python
print(titanic_data['sex'].unique())
print(titanic_data['embark_town'].unique())
print(titanic_data['class'].unique())
```

```python
import pandas as pd 
temp = pd.get_dummies(titanic_data['sex'])

temp.head()
```

```python
pd.concat([titanic_data['sex'], temp],axis=1).head()
```

```python
temp = pd.get_dummies(titanic_data['embark_town'], dummy_na=True, drop_first=True)
pd.concat([titanic_data['embark_town'],temp],axis=1).head()

```

3.3.2 Label encoding

categories are represented as integers

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(titanic_data['class'])

titanic_data['le_class'] = le.transform(titanic_data['class'])

titanic_data.head(10)

```
